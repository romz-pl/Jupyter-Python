## Letâ€™s draw sine and cosine functions using Matplotlib defaults.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

## Letâ€™s now redo the figure but with the specification of all the different settings.

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
p = dict()
p["figure.figsize"] = 6,2.5
p["figure.dpi"] = 100
p["figure.edgecolor"] = "black"
p["figure.facecolor"] = "#f9f9f9"
p["axes.linewidth"] = 1
p["axes.facecolor"] = "#f9f9f9"
p["axes.xmargin"] = 0.1
p["axes.ymargin"] = 0.1
p["axes.spines.bottom"] = True
p["axes.spines.left"] = True
p["axes.spines.right"] = False
p["axes.spines.top"] = False
p["font.sans-serif"] = ["Fira Sans Condensed"]
p["axes.grid"] = False
p["grid.color"] = "black"
p["grid.linewidth"] = .1
p["xtick.bottom"] = True
p["xtick.top"] = False
p["xtick.direction"] = "out"
p["xtick.major.size"] = 5
p["xtick.major.width"] = 1
p["xtick.minor.size"] = 3
p["xtick.minor.width"] = .5
p["xtick.minor.visible"] = True
p["ytick.left"] = True
p["ytick.right"] = False
p["ytick.direction"] = "out"
p["ytick.major.size"] = 5
p["ytick.major.width"] = 1
p["ytick.minor.size"] = 3
p["ytick.minor.width"] = .5
p["ytick.minor.visible"] = True
p["lines.linewidth"] = 2
p["lines.markersize"] = 5
p["lines.linestyle"] = "solid"

fig = plt.figure(figsize = p['figure.figsize'], dpi = p['figure.dpi'], facecolor = p['figure.facecolor'], edgecolor = p['figure.edgecolor'])
ax = plt.subplot(1,1,1)
ax.plot(X, C, color="C0", linewidth = p['lines.linewidth'], linestyle = p['lines.linestyle'])
ax.plot(X, S, color="C1", linewidth = p['lines.linewidth'], linestyle = p['lines.linestyle'])
xmin, xmax = X.min(), X.max()
xmargin = p['axes.xmargin']*(xmax - xmin)
ax.set_xlim(xmin - xmargin, xmax + xmargin)
ymin, ymax = min(C.min(), S.min()), max(C.max(), S.max())
ymargin = p['axes.ymargin']*(ymax - ymin)
ax.set_ylim(ymin - ymargin, ymax + ymargin)
ax.tick_params(axis = "x", which="major", direction = p['xtick.direction'], length = p['xtick.major.size'], width = p['xtick.major.width'])
ax.tick_params(axis = "y", which="major", direction = p['ytick.direction'], length = p['ytick.major.size'], width = p['ytick.major.width'])
plt.show()
```

## Vertical lines with various colours

```python
fig, ax = plt.figure(figsize = (8,5)), plt.axes()
x = np.linspace(0,100,2)
ax.plot(x, x, color = 'gray')
ax.set_xlim([0,100])
ax.set_ylim([0,100])

for i in range(101):
    ax.axvline(i, 0    , i/100, color = 'C' + str(i))
    ax.axvline(i, i/100, 1    , color = 'C' + str(i+5))

ax.axis('off')
plt.savefig('./output/colorful.pdf')
plt.show()
```

## Import the iris dataset and make a boxplot

```python
data = load_iris()['data']
# print(data)
df = pd.DataFrame(data)
fig = plt.figure()
ax = plt.axes()
df.plot.box(ax = ax)
ax.yaxis.grid(True)
ax.xaxis.grid(False)
plt.tight_layout()
plt.savefig('./output/irisbox.pdf')
plt.show()
```

## Set the limits on the axes

```python
fig, ax = plt.figure(), plt.axes()

for i in range(1,4):
    ax.plot([0,i], [i,i])
    bottom_y, top_y = ax.get_ylim()
    left_x, right_x = ax.get_xlim()
    ax.fill_between(x = [left_x,right_x], y1 = bottom_y, y2 = top_y, alpha = 0.5/i)

ax.set_ylim(bottom_y, top_y)
ax.set_xlim(left_x, right_x)
plt.show()
```

## Spine modifications

```python
fig, ax = plt.figure(), plt.axes(facecolor = 'lightyellow')
ax.set_title("Thick Spines")
for spine in 'bottom', 'top', 'left', 'right':
    ax.spines[spine].set_color('black')
    ax.spines[spine].set_linewidth(4)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
plt.show()
```

## Use LATEX strings to label the ticks

```python
x = np.linspace(0, np.pi * 2, 100)
fig, ax = plt.figure(), plt.axes()
ax.plot(x, np.sin(x))
ax.set_yticks( [-0.5, 0, 0.5] )
ax.set_yticklabels( [r"$-\frac{1}{2}$", 0,  r"$\frac{1}{2}$"] )
ax.xaxis.set_ticks([np.pi/2, np.pi])
ax.xaxis.set_ticklabels([r"$\frac{\pi}{2}$", r"$\pi$"])
plt.show()
```

## Swoh Pandas.Series as bar plot 

```python
data = pd.Series( {'Shaq': 7 + (1/12),
                     'Yao Ming': 7.5,
                     'Delle Donne': 6 + (5/12)})

fig, ax = plt.figure(figsize = (4,7)), plt.axes()

data.plot.bar(ax = ax,
        color = ['#FDB927', '#BA0C2F', '#0C2340'],
        edgecolor = ['#552583', '#041E42', '#C8102E'],
        linewidth = 2)

# Get rid of ticks on x-axis, rotate text
ax.xaxis.set_tick_params(length = 0, which = 'major', rotation = 0)

ylim0, ylim1 = 0,8
ax.set_ylim([ylim0, ylim1])

ax.set_yticks(range(ylim0, ylim1+1))

# ax.yaxis.set_major_locator(tck.MultipleLocator(1))
ax.yaxis.set_minor_locator(tck.MultipleLocator(1/12))
ax.yaxis.set_tick_params(length = 1, which = 'minor')
ax.yaxis.set_tick_params(length = 5, which = 'major')

ax.set_ylabel("Height (feet)")
ax.set_title("Pro Basketball Players are Tall")
plt.show()
```
